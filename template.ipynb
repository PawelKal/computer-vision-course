###Wycinanie fragmentów obrazów

```
img = cv2.imread(filename='view.jpg')
cv2_imshow(img)
```

```
left_top_square = img[:200, :200]
print(left_top_square.shape)
cv2_imshow(left_top_square)
```

```
right_top_square = img[:200,-200:]
print(right_top_square.shape)
cv2_imshow(right_top_square)
```

```
left_bottom_square = img[-200:,:200]
cv2_imshow(left_bottom_square)
```

```
right_bottom_square = img[-200:,-200:]
cv2_imshow(right_bottom_square)
```

```
height, width, _ =img.shape

center_height = height//2
center_width = width//2

img_heigiht = 100
img_width = 100

middle = img[center_height-img_heigiht:center_height+img_heigiht, center_width-img_width:center_width+img_width]
cv2_imshow(middle)
```

###Modyfikowanie obrazu

####Ramka

```
replicate = cv2.copyMakeBorder(src=img, top=50, bottom=50, left=50, right=50, borderType=cv2.BORDER_REPLICATE)
cv2_imshow(replicate)
```

```
replicate = cv2.copyMakeBorder(src=img, top=50, bottom=50, left=50, right=50, borderType=cv2.BORDER_REFLECT)
cv2_imshow(replicate)
```
